# Seaborn Visualization

```python
# univariate analysis of Internet users
vs1 = sns.distplot(df['InternetUsers'])
```

```python
# Bivariate analysis using boxplot
vs2 = sns.boxplot(data=df,x='IncomeGroup',y='BirthRate')
```
