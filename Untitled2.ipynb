# MAT-PLOTLIB PRACTICE

```python
# importing libraries
import matplotlib.pyplot as plt
import numpy as np
```

```python
# generating x-values
x_values = np.linspace(-5,5,50)
y_values = np.sin(x_values)
z_values = np.cos(x_values)

#plotting the graphs
plt.plot(x_values,y_values,'r') #The graph for sine
plt.plot(x_values,z_values,'g') #The graph for cosine
# setting labels
plt.xlabel('Angles')
plt.ylabel('Sin(Angle)')
plt.title('T-ratios Graph')
plt.show()
```

```python
# Plotting Linear graph of y = -x + 1
plt.plot(x_values,-x_values+1,'r.')
plt.plot(x_values,x_values+1,'r.')
# fig = plt.figure()
# axis= fig.add_axes([0,0,1,1])
# axis.scatter(x_values,-x_values+1)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
# plotting the parabolic curve
plt.plot(x_values,x_values**2)
plt.show()
```

```python
# datasets
subjects = np.array(['Math','Physics','Biology','Geography'])
scores = np.array([67,67,100,89])
# plotting the bar graph
fig = plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.bar(subjects,scores)
plt.xlabel('Subjects')
plt.ylabel('Scores')
plt.title('Scores against Subjects')
plt.show()
```

```python
# Plotting pie chart
fig2 = plt.figure()
axes = fig2.add_axes([0,0,1,1])
axes.pie(scores,labels=subjects,autopct='%.2f%%')
plt.show()
```

```python
# creating scatter graph
fig3 = plt.figure()
axes = fig3.add_axes([0,0,1,1])
axes.scatter(x_values,y_values)
axes.scatter(x_values,z_values)
plt.show()
```

```python
# datasets
z=np.linspace(-5,5,50)
x=np.sin(z)
y=np.cos(z)
# plotting 3D scatter
axes = plt.axes(projection='3d')
axes.scatter(x,y,z,c=z,cmap='Blues')
# axes.scatter(x,y,z)
plt.show()
```
