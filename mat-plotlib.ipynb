# MAT-PLOTLIB PRACTICE

```python
# import libraries
import matplotlib.pyplot as plt
import numpy as np
```

```python
# creating evening spaced range of values from 0-20
x_values = np.linspace(-0,5,100)
x_values
```

```python
# Creating the identical function
y_values = np.sin(x_values)
y_values
```

```python
# creating the z-values for cosine function
z_values = np.cos(x_values)
z_values
```

```python
# plotting the sine graph
plt.plot(x_values,y_values,'gx') #sine function
plt.plot(x_values,z_values,'rx') #cosine function
# plt.plot(x_values,x_values**2)
plt.xlabel('Angles')
plt.ylabel('Output')
plt.title('Sine Function')
plt.show()
```

```python
# Plotting the cosine graph
plt.plot(x_values,z_values,'bx')
plt.xlabel("Angles")
plt.ylabel('Output')
plt.title('Cosine Function')
plt.show()
```

```python
# creating the parabolic curve
plt.plot(x_values,x_values + 100,'r.')
plt.xlabel('Students')
plt.ylabel('Grade')
plt.title('Grade against Students')
plt.show()
```

```python
# plotting the bar graph
# making the figure instance
fig = plt.figure()
axis =fig.add_axes([0,0,1,1])
subjects = np.array(['English','Math','Biology','Geography'])
scores   = np.array([80,97,60,79])
# plotting
axis.bar(subjects,scores)
# Adding labels
plt.xlabel('Subjects')
plt.ylabel('Scores')
plt.title('Subjects Against Scores')
```

```python
# creating a pie chart
fig1 = plt.figure()
axis = fig1.add_axes([0,0,1,1])
axis.pie(scores,labels=subjects,autopct='%.2f%%')
```
